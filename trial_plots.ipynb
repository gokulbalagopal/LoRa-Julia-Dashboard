```julia
using CSV
using DataFrames
using Dates
using Plots
plotly()

df = CSV.read("47cb5580002e004a_2021-06-17.csv",DataFrame)
df.dateTime = DateTime.(df.dateTime,"yyyy-mm-dd HH:MM:SS")
hs = [Plots.plot(df.dateTime, df[:,i], title=names(df)[i]) for i in 4:ncol(df)]
ncols = 5 # set how many columns you want in the plot
nrows = cld(length(hs), ncols)
blankplot = Plots.plot(legend=false,grid=false,foreground_color_subplot=:white)
for i = (length(hs)+1):(nrows*ncols)
    push!(hs, deepcopy(blankplot))
end
Plots.plot(hs..., layout=(nrows, ncols), legend=false, size=(3000,1500))
```
